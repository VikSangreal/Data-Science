# Height

```python
df['Height']
df= df[pd.notnull(df['Height'])]
def pr(h):
  f=h.split("'")
  g=float(f[0])
  h1=float(f[1].replace("\"",""))
  return(12*g)+h1                    
df['Height']=(df['Height'].map(lambda x: pr(x)))
df['Height']
```

# Wage 

```python
df['Wage'].astype(str)
df= df[pd.notnull(df['Wage'])]
def f(x):
 for i in x:
   if(i=='K'):
        x=x.rstrip(i)
        x=float(x)
        x=x*1000
        return x
   elif(i=='M'):

        x=x.rstrip(i)
        x=float(x)
        x=x*1000000
        return x
   
   
   
        
df['Wage']=df['Wage'].str.lstrip('€')
df['Wage']=df['Wage'].apply(lambda x: f(x))
df['Wage']=df['Wage'].astype(float)
df['Wage']
```

# Value

```python
df['Value']=df['Value'].str.lstrip('€')
df['Value']=df['Value'].apply(lambda x: f(x))
#df['Value']=df['Value'].str.rstrip('M')
df['Value']=df['Value'].astype(float)
df['Value']
```

# Release Clause

```python
def f(x):
 x=str(x)
 for i in x:
   if(i=='K'):
        x=x.rstrip(i)
        x=float(x)
        x=x*1000
        return x
   elif(i=='M'):

        x=x.rstrip(i)
        x=float(x)
        x=x*1000000
        return x
df['Release Clause']=df['Release Clause'].str.lstrip('€')
df['Release Clause']=df['Release Clause'].apply(lambda x: f(x))
#df['Release Clause']=df['Release Clause'].str.rstrip('M')
df['Release Clause']=df['Release Clause'].astype(float)
df['Release Clause']
```

# Weight

```python
df= df[pd.notnull(df['Weight'])]
df['Weight']=df['Weight'].str.rstrip('lbs')
df['Weight']=df['Weight'].astype(float)
df['Weight']
```

# Joined

```python
df= df[pd.notnull(df['Joined'])]
def pr(h):
  f=h.split(",")
  g=int(f[1])
  return g                 
df['Joined']=(df['Joined'].map(lambda x: pr(x)))
df['Joined']
```

# Contract Valid Until

```python
df['Contract Valid Until']=pd.to_datetime(df['Contract Valid Until'])
df['Contract Valid Until']=df['Contract Valid Until'].dt.year
```
